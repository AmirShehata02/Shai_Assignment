# Some Cleaning

```python
predata.info()
```

```python
predata.rename(columns = {'TotalPayBenefits':'TotalIncome'}, inplace = True)
predata.rename(columns = {'TotalPay':'Salary'}, inplace = True)  
predata.head()           
```

```python
predata['Department'] = predata['JobTitle'].str.extract(r'\((.*?)\)')

# Display the updated DataFrame
predata['Department']
```

# Analysis

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6, 4))

sns.histplot(predata.Salary, bins=40)
plt.ylabel('Frequency')
plt.xlabel('Salary')
plt.title('Distribution of prices')

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(predata.Year, bins=40,color='blue', edgecolor='black')
plt.ylabel('TotalIncome')
plt.xlabel('Year')
plt.title('Distribution of Income')

plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.histplot(predata.JobTitle, bins=40,color='blue', edgecolor='black')
plt.ylabel('TotalIncome')
plt.xlabel('JobTitle')
plt.title('Distribution of Income')

plt.show()
```

```python
Count_Employee = predata.shape[0]
Department_Counts = predata['Department'].value_counts()
plt.figure(figsize=(8, 8))
plt.figure(figsize=(8, 8))
plt.pie(Department_Counts, labels=Department_Counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue', 'lightgreen', 'gold', 'orchid'])
plt.title('Employee Distribution by Department')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# If you're encountering an error with len(), you might want to convert the float to a scalar
try:
    plt.show()
except TypeError:
    plt.show().item()
    
```

```python
predata
```

```python
Grouped_Data_1 = predata.groupby(['Department','JobTitle']).agg({'Salary': 'mean'}).reset_index()
Grouped_Data_1
```

```python
Correlation = predata['Salary'].corr(predata['OvertimePay'])

# Print the correlation coefficient
print(f"Correlation between Salary and OvertimePay: {Correlation}")

# Plot a scatter plot to visualize the relationship
plt.scatter(predata['OvertimePay'], predata['Salary'])
plt.title('Scatter Plot: Salary vs OvertimePay')
plt.xlabel('OvertimePay')
plt.ylabel('Salary')
plt.show()
```
